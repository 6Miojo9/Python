```python
import os
import pandas as pd           

# Defina o caminho da pasta onde está a planilha
pasta = 'G01 0209'

# Listar todos os arquivos na pasta
arquivos = os.listdir(pasta)

planilhas = [arquivo for arquivo in arquivos if arquivo.endswith('.xlsx')]

# Verificar se há planilhas e selecionar a primeira (ou aplicar outra lógica de seleção)
if planilhas:
    caminho_planilha = os.path.join(pasta, planilhas[0])
    
    df = pd.read_excel(caminho_planilha)
    print(df)
else:
    print('Nenhuma planilha encontrada.')
    print()
```
